```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://mydramalist.com/search?q=The+K2"
```

```python
HEADERS = ({'User=Agent':'https://explore.whatismybrowser.com/useragents/parse/?analyse-my-user-agent=yes','Accept-Language':'en-US, en;q=0.5'})
```

```python
page = requests.get(url, headers=HEADERS)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
soup
```

```python
links = soup.find_all("div", attrs={'class':'col-xs-9 row-cell content'})
links
```

```python
links[0]
```

```python
links = soup.find("div", attrs={'class':'col-xs-9 row-cell content'}).find_all('a')
links
```

```python
link_href = links[0].get('href')
link_href
```

```python
drama_list = "https://mydramalist.com" + link_href
drama_list
```

```python
new_page = requests.get(drama_list, headers=HEADERS)
new_page
```

```python
new_soup = BeautifulSoup(new_page.content, "html.parser")
new_soup
```

```python
drama_element = soup.find('h6', class_='text-primary title').find('a')
drama_element
```

```python
drama_name = drama_element.get_text()
drama_name
```

```python
rating_element = soup.find('span', class_='p-l-xs score')
rating_element
```

```python
rating = float(rating_element.get_text())
rating
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
plt.barh([drama_name], [rating], color='skyblue')
plt.xlabel('Rating')
plt.ylabel('Drama Name')
plt.title('Rating Visualization')
plt.xlim(0, 10)
plt.show()
```
